```r
library(ggplot2) #plotting library
data = read.csv("kaggle_data_0228.csv") #enter data file name (make sure the file is in the same directory as your notebook)
team_data= data[which(data$Team.Name=="BAH_natomy"),] #pulls our team data
score_data= data[which(data$Score < 1.1),] #removes extreme scores from data to make plot more readable
```

```r
#extracts important data columns for plotting
place=data[,"Num"]
score=data[,"Score"]
place1=score_data[,"Num"]
score1=score_data[,"Score"]

#figures out place required for bronze, silver, and gold
qt_gold= ceiling((nrow(data)*.02) + 10)
qt_bronze=ceiling(nrow(data)*.10)
qt_silver=ceiling(nrow(data)*.05 )
```

```r
#creates plot and identifies our team's location
plot(place1, score1, main="Scatterplot of DSB 2017 Leader Board", col=ifelse(data$Team.Name=="BAH_natomy","dodgerblue2","gray40"),pch=ifelse(data$Team.Name=="BAH_natomy",24,20),cex=ifelse(data$Team.Name=="BAH_natomy",2,1), bg="dodgerblue2", xlab="Place", ylab="Score")

#draws lines to represent score break out for gold, silver, and bronze
abline(v = qt_bronze, col = "darkorange3",lwd=2 )
abline(v = qt_silver, col = "cornsilk4",lwd=2 )
abline(v = qt_gold, col = "darkgoldenrod",lwd=2)

#creates variables for plot text
gold= paste("gold =", qt_gold)
silver= paste("silver =", qt_silver)
bronze= paste("bronze =", qt_bronze)
team_place = paste("BAH_natomy: ", team_data[1,1],",", team_data[1,4])

#finds current date and how many days left to the dealines
currdate= Sys.Date()
deadline= as.Date("2017-04-07")
deadline2= as.Date("2017-04-12")
timeleft= difftime(deadline,currdate,units = c("days"))
timeleft2= difftime(deadline2,currdate,units = c("days")) 
currdate= paste(format(currdate, format="%b %d %Y"))
plottimestamp= paste(currdate,", Stage 1 Deadline:",timeleft,"days", ", Final Submission Deadline:", timeleft2,"days")

#adds text to the plot
text(1000,.08,gold, adj=0)
text(1000,.04,silver, adj=0)
text(1000,0,bronze, adj=0)
text(400,.5,team_place, adj=0)
mtext(plottimestamp, side = 3)
```
